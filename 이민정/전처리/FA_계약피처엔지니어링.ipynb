```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv('KBO FA 계약 내역(2013-2024).csv의 사본.csv', encoding='cp949')
df.head(15)
```

```python
df['1년당_FA금액'] = df['FA 계약 총액'] / df['계약 연수']
```

```python
a1 = df.groupby(['FA Year','잔류 여부']).size().unstack(fill_value=0)
a1 = a1.reset_index()
a1.columns.name = None
a1
```

```python
a1['이적률'] = a1['이적'] / ( a1['이적']+a1['잔류'] ) * 100
```

```python
a2 = df.groupby(['FA Year', '잔류 여부'])[['계약 연수', 'FA 계약 총액']].sum().reset_index()
a2['1년당_FA금액'] = a2['FA 계약 총액']/ a2['계약 연수']

a2
```

```python
result = result = pd.merge(a1, a2, on='FA Year', how='inner')
result
```

```python
result['전년_1년당_FA금액'] = result.groupby('잔류 여부')['1년당_FA금액'].shift(1)
result['전년_대비_증감률'] = ((result['1년당_FA금액'] - result['전년_1년당_FA금액']) / result['전년_1년당_FA금액']) * 100
```

```python
result = result.drop(['전년_1년당_FA금액'], axis=1)
```

```python
result = result.fillna(0)
result = result.round(2)
result.columns = [['FA_Year', '이적', '잔류', '이적률', '잔류_여부', '총_계약_연수','총_FA_금액','1년당_FA금액','전년_대비_증감률']]
```

```python
result.to_csv('년도별FA정보.csv', index = False)
```
