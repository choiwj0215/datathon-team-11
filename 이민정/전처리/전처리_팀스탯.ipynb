```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('0801_야수_팀스탯.csv')
df
```

```python
### 태블로에서 누적막대그래프를 만들기 위해 스케일링
```

```python
df.columns
```

```python
stats_cols = ['WAR', '게임 수', 'oWAR', 'dWAR', '타석', '실질타석',
       '타수', '득점', '안타', '2루타', '3루타', '홈런', '루타', '타점', '도루성공', '도루실패', '볼넷',
       '사구', '고의사구', '삼진', '병살타', '희생번트', '희생플라이', '타율', '출루율', '장타율', 'OPS',
       '득점/실질타석', 'wRC+', '타격WPA', '주루WPA', 'WPA+', 'WPA-', 'WPA']

scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[stats_cols] = scaler.fit_transform(df[stats_cols])
```

```python
df_scaled
```

```python
# 저장
df_scaled.to_csv('0801_야수_팀스탯_정규화ver.csv', index=False)
```
