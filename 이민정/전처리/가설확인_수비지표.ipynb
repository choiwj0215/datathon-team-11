```python
import pandas as pd
import statsmodels.api as sm
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("C:/Users/이민정/Desktop/데이터톤/data/스탯티즈_Raw_data/FA_야수데이터_수정전.csv")
df
```

```python
df_clean = df.dropna(subset=['평균대비수비득점기여', '수비_포지션 RAA', '전체수비 RAA', '수비WAR'])
```

```python
df_clean['FA 계약 총액']= df_clean['FA 계약 총액'].fillna(0)
```

```python
correlation = df_clean[['수비_포지션 RAA','평균대비수비득점기여','전체수비 RAA','수비WAR','연봉', 'FA 계약 총액']].corr()
correlation
```

```python
X = df_clean[['수비WAR', '전체수비 RAA', '수비_포지션 RAA', '평균대비수비득점기여']]
y = df_clean['FA 계약 총액']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
X = df_clean[['수비WAR']]
y = df_clean['FA 계약 총액']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
