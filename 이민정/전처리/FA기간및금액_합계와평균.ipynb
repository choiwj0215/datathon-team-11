```python
import pandas as pd
```

```python
df = pd.read_excel('FA_계약현황_구단_기간_총액.xlsx')
df
```

```python
df['계약기간'] = df['계약기간'].str.extract(r'(\d+)').astype(int)

df['총액'] = df['총액'].str.replace('억', '', regex=False)
df['총액'] = df['총액'].astype(float).astype(int)
```

```python
df
```

```python
summary_all_teams = df.groupby('구단명').agg({
    '계약기간': ['sum', 'mean'],
    '총액': ['sum', 'mean']
}).reset_index()

summary_all_teams.columns = ['구단명', '계약기간_합계', '계약기간_평균', '총액_합계', '총액_평균']

summary_all_teams
```

```python
summary_by_year = df.groupby('계약년도').agg({
    '계약기간': ['sum', 'mean'],
    '총액': ['sum', 'mean']
}).reset_index()

summary_by_year.columns = ['계약년도', '계약기간_합계', '계약기간_평균', '총액_합계', '총액_평균']

total = pd.DataFrame({
    '계약년도': ['전체'],
    '계약기간_합계': [df['계약기간'].sum()],
    '계약기간_평균': [df['계약기간'].mean()],
    '총액_합계': [df['총액'].sum()],
    '총액_평균': [df['총액'].mean()]
})

summary_by_year = pd.concat([summary_by_year, total], ignore_index=True)
summary_by_year
```

```python
summary_team_year = df.groupby(['구단명', '계약년도']).agg({
    '계약기간': ['sum', 'mean'],
    '총액': ['sum', 'mean']
}).reset_index()

summary_team_year.columns = ['구단명', '계약년도', '계약기간_합계', '계약기간_평균', '총액_합계', '총액_평균']

summary_team_year
```

```python
summary_all_teams.to_excel('구단별_FA계약.xlsx', index = False)
summary_team_year.to_excel('구단,년도별_FA계약.xlsx', index = False)
```
