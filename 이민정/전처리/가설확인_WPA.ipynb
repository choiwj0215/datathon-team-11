```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib import rc
import platform

import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)

# 한글 폰트 설정
plt.rc('font', family='Malgun Gothic')

# 마이너스 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('야수_전처리_ver_1년차제거.csv')
df
```

```python
df[['타격 WPA','주루 WPA','WPA+','WPA-','WPA', 'FA 계약 총액']].corr()
```

```python
df1 = df.copy()
df1 = df1.dropna()
```

```python
X = df1[['타격 WPA','주루 WPA','WPA+','WPA-','WPA']]
X = sm.add_constant(X)
y = df1['FA 계약 총액']

model = sm.OLS(y, X).fit()
print(model.summary())
```
