```python
import pandas as pd
```

```python
df = pd.read_csv('FA_야수_수정데이터.csv')
df
```

```python
df.columns
```

```python
df = df.drop(['1년당_계약금액'],axis=1)
```

```python
nan_counts = df.isna().sum()
nan_counts = nan_counts[nan_counts > 0]
print(nan_counts)
```

```python
df['계약기간'] = df['계약기간'].str.replace('년', '', regex=False).astype(float)
df['총액'] = df['총액'].str.replace('억', '', regex=False).astype(float)
```

```python
result = df.groupby('선수명').agg({
    '계약기간': 'sum',
    '총액': 'sum'
}).reset_index()

result
```

```python
result['1년당_계약금액'] = result['총액'] / result['계약기간']
```

```python
result
```

```python
result1 = result.drop(['계약기간','총액'], axis= 1)
result1
```

```python
merged = pd.merge(df, result1, on='선수명', how='left')

merged
```

```python
merged.to_csv('FA_야수데이터.csv',encoding='utf-8-sig', index=False)
```
