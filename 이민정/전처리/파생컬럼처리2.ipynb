```python
import pandas as pd
from collections import Counter
pd.set_option('display.max_columns', None)
```

```python
a1 = pd.read_csv('야수_전처리_ver05.csv')
df = pd.read_excel('국가대표명단.xlsx')
a1
```

```python
#띄어쓰기 제거 (문자열 컬럼 전체에 대해)
for col in df.columns:
    if df[col].dtype == object:
        df[col] = df[col].str.replace(' ', '', regex=False)
```

```python
#이름 컬럼들 찾기
name_cols = [col for col in df.columns if '이름' in col]

#모든 이름을 모아서 리스트로 만들기
all_names = []

for col in name_cols:
    all_names.extend(df[col].dropna().tolist())
```

```python
#선수별 출전 횟수 계산
counter = Counter(all_names)

#출전 횟수에 따라 여러번/한번 출전 선수 분리
once_players = {name: cnt for name, cnt in counter.items() if cnt == 1}
multi_players = {name: cnt for name, cnt in counter.items() if cnt > 1}

#데이터프레임 생성
df_once = pd.DataFrame({
    '한번출전한선수': list(once_players.keys()),
    '출전횟수': list(once_players.values())
})

df_multi = pd.DataFrame({
    '여러번출전한선수': list(multi_players.keys()),
    '여러번출전횟수': list(multi_players.values())
})
```

```python
#두 데이터프레임 병합
df1 = pd.merge(df_once, df_multi, left_index=True, right_index=True, how='outer')

df1 = df1.fillna('')
df1 = df1.reset_index(drop=True)
```

```python
df1.to_csv('국가대표_출전횟수.csv', index= False)
df1.to_excel('국가대표_출전횟수.xlsx', index= False)
```
