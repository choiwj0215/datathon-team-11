```python
import pandas as pd
```

```python
a1 = pd.read_csv('야수_전처리_ver05.csv')
df1 = pd.read_csv('국대1번출전선수.csv')
df2 = pd.read_csv('국대여러번출전선수.csv')
```

```python
a1['선수명_수정'] = a1['선수명'].str.replace(r'\(\d+\)', '', regex=True)
a1
```

```python
# 선수명 기준으로 merge
merged = pd.merge(a1, df1, left_on='선수명_수정', right_on='선수명', how='left')
merged
```

```python
merged = merged.drop(columns=['선수명_수정', '선수명_y'])
merged = merged.rename(columns={'선수명_x':'선수명'})

merged
```

```python
merged.to_csv('야수_전처리_ver06.csv',index=False)
```

```python
a2 = pd.read_csv('야수_전처리_ver06.csv')
```

```python
a2['선수명_수정'] = a2['선수명'].str.replace(r'\(\d+\)', '', regex=True)
a2
```

```python
# 선수명 기준으로 merge
merged1 = pd.merge(a2, df2, left_on='선수명_수정', right_on='선수명', how='left')
merged1
```

```python
merged1 = merged1.drop(columns=['선수명_수정', '선수명_y'])
merged1 = merged1.rename(columns={'선수명_x':'선수명'})

merged1
```

```python
merged1['국대출전횟수'] = merged1['국대출전횟수_x'].combine_first(merged1['국대출전횟수_y'])
```

```python
merged1
```

```python
merged1 = merged1.drop(['국대출전횟수_x','국대출전횟수_y'],axis=1)
```

```python
merged1
```

```python
merged1['국대출전횟수']= merged1['국대출전횟수'].fillna(0)
```
