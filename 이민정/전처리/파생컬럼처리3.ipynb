```python
import pandas as pd
```

```python
a1 = pd.read_csv('야수_전처리_ver06.csv')
df1 = pd.read_csv('국대1번출전선수.csv')
df2 = pd.read_csv('국대여러번출전선수.csv')
```

```python
a1['선수명_수정'] = a1['선수명'].str.replace(r'\(\d+\)', '', regex=True)
a1
```

```python
# 선수명 기준으로 merge
merged = pd.merge(a1, df2, left_on='선수명_수정', right_on='선수명', how='left')
merged
```

```python
merged = merged.drop(columns=['선수명_수정', '선수명_y'])
merged = merged.rename(columns={'선수명_x':'선수명'})

merged
```

```python
merged['국대출전횟수'] = merged['국대출전횟수_x'].combine_first(merged['국대출전횟수_y'])
```
