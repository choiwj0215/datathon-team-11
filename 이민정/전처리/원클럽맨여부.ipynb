```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('야수_전처리_ver07.csv')
df
```

```python
# 선수명만 남게 컬럼을 만든다
df['선수명_수정'] = df['선수명'].str.replace(r'\(\d{4}\)', '', regex=True)
```

```python
# 잔류여부가 이적이나 트레이드인 경우 0을 넣어준다
df['원클럽맨여부'] = np.where(df['잔류 여부'].isin(['이적', '트레이드']), 0, np.nan)
```

```python
# 잘 적용 되었는지 선수 몇명 검색해서 확인
df[df['선수명'].str.contains(r'^김선빈', regex=True)]
```

```python
# 이대호 선수는 수기로 수정해기기
df.loc[[390,391,392,393],'원클럽맨여부']=1
```

```python
# 원클럽맨여부에 0이 들어간 선수들 모두 0으로 채워넣기
names_zero = df.loc[df['원클럽맨여부'] == 0, '선수명_수정'].unique()
df.loc[df['선수명_수정'].isin(names_zero), '원클럽맨여부'] = 0
```

```python
names = df[df['원클럽맨여부'].isna()]['선수명_수정'].drop_duplicates()
names
```

```python
a1 = pd.read_excel('원클럽맨여부.xlsx')
a1
```

```python
df['원클럽맨여부'] = df['선수명_수정'].map(a1.set_index('선수명')['원클럽맨여부'])
```

```python
names_one = df.loc[df['원클럽맨여부'] == 1, '선수명_수정'].unique()
df.loc[df['선수명_수정'].isin(names_one), '원클럽맨여부'] = 1
```

```python
df = df.drop(['선수명_수정'],axis=1)
df
```
