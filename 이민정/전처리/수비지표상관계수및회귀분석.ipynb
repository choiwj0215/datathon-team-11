```python
import pandas as pd
import statsmodels.api as sm
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('야수_전처리_ver03_데이터추가.csv')
df
```

```python
df_clean = df.dropna(subset=['평균대비수비득점기여', '수비_포지션 RAA', '전체수비 RAA', '수비WAR'])
```

```python
grouped = df.groupby('선수명')[['공격WAR','wRC+','장타율','출루율','OPS',
                             '수비_포지션 RAA','평균대비수비득점기여','전체수비 RAA','수비WAR',
                             '연봉', 'FA 계약 총액']].sum().reset_index()
```

```python
correlation = grouped[['수비_포지션 RAA','평균대비수비득점기여','전체수비 RAA','수비WAR','연봉', 'FA 계약 총액']].corr()
correlation
```

```python
correlation = grouped[['공격WAR','wRC+','장타율','출루율','OPS',
                             '수비_포지션 RAA','평균대비수비득점기여','전체수비 RAA','수비WAR',
                             '연봉', 'FA 계약 총액']].corr()
correlation
```

```python
X = grouped[['수비WAR', '전체수비 RAA', '수비_포지션 RAA', '평균대비수비득점기여']]
y = grouped['연봉']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
