```python
import pandas as pd
```

```python
df = pd.read_excel('무사주자1루상황_2013~2025.xlsx')
df1 = pd.read_excel('1사주자2루상황_2013~2025.xlsx')
```

```python
no_out = df[df['팀']=='전체']
no_out
```

```python
one_out = df1[df1['팀']=='전체']
one_out.columns
```

```python
no_out = no_out[['팀', '년도', '타석', '타수', '득점', '안타', '2루타', '3루타', '홈런', '타점',
       '볼넷', '사구', '고의4구','병살타', '희생번트', '타율', '출루율', '장타율', 'OPS','순장타율', '순출루율', 'BABIP']]
```

```python
one_out = one_out[['팀', '년도', '타석', '타수', '득점', '안타', '2루타', '3루타', '홈런', '타점',
       '볼넷', '사구', '고의4구','병살타', '희생번트', '타율', '출루율', '장타율', 'OPS','순장타율', '순출루율', 'BABIP']]
```

```python
merged = pd.merge(no_out, one_out,on=['년도'])
merged
```

```python
merged.columns = merged.columns.str.replace('_x', '_무사').str.replace('_y', '_1사')
```
