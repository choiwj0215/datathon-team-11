### 저장한 모델 객체 등을 복원한다

```python
best_model_path = 'model/best_model_regression_choice_fa_2024_KNN_tuning.dat'

with open(best_model_path, 'rb') as fp :
    model1 = pickle.load(fp)
    scaler1 = pickle.load(fp)

display(model1)
display(scaler1)
```

### 예측할 데이터를 준비한다.

- 모든 준비가 완료되면 입력데이터는 test_X 라는 변수에 담아주도록 한다.

```python
# 데이터를 불러온다.
df1 = pd.read_csv('머신러닝_2026_FA예정_타자.csv')
df1
```

```python
feature_columns = list(scaler1.feature_names_in_)

for col in feature_columns:
    if col not in df1.columns:
        df1[col] = 0

# 컬럼 순서도 학습 때와 동일하게 맞춤
df1 = df1[feature_columns]
```

```python
# 전처리
scaled_data = scaler1.transform(df1)
scaled_data
```

```python
test_X = scaled_data
```

```python
10/0
```

### 예측하고 저장한다

```python
# 예측한다.
y_pred = model1.predict(test_X)
y_pred
```

```python
# 결과를 붙힌다음 저장한다.
df1['1년당 계약 총액'] = y_pred
df1.to_csv(prediction_path, index=False, encoding='utf-8-sig')
print('저장완료')
```
