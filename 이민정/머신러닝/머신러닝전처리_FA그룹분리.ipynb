```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('2013~2016_FA_4년정보.csv')
df1 = pd.read_csv('2017~2024_FA_4년정보.csv')
df1
```

```python
df1.columns
```

```python
df1[df1['선수명_수정']=='이성열']
```

```python
# 1년당 계약 총액 계산: 계약 총액 / 계약 연수
df['1년당 계약 총액'] = df['FA 계약 총액'] / df['FA 계약 연수']

# 선수명 기준으로 마지막 행(=유일하게 계약총액이 있는 행)의 값을 전체에 채움
df['1년당 계약 총액'] = df.groupby('선수명')['1년당 계약 총액'].transform('max')
```

```python
# 1년당 계약 총액 계산: 계약 총액 / 계약 연수
df1['1년당 계약 총액'] = df1['FA 계약 총액'] / df1['FA 계약 연수']

# 선수명 기준으로 마지막 행(=유일하게 계약총액이 있는 행)의 값을 전체에 채움
df1['1년당 계약 총액'] = df1.groupby('선수명')['1년당 계약 총액'].transform('max')
```

```python
null = df.isna().sum()
null = null[null>0]
null
```

```python
df = df.drop(['구단명', '선수명','FA 계약 연수','FA 계약 총액','선수명_수정','잔류 여부'], axis = 1)
df1 = df1.drop(['구단명', '선수명','FA 계약 연수','FA 계약 총액','선수명_수정','잔류 여부'], axis = 1)
```

```python
df['메인 타순'] = df['메인 타순'].str.extract('(\d+)').astype(float)
df1['메인 타순'] = df1['메인 타순'].str.extract('(\d+)').astype(float)
```

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_ohe = ohe.fit_transform(df[categorical_cols])

ohe_cols = ohe.get_feature_names_out(categorical_cols)

df_ohe = pd.DataFrame(X_ohe, columns=ohe_cols, index=df.index)

df_final = pd.concat([df.drop(columns=categorical_cols), df_ohe], axis=1)
```

```python
categorical_cols1 = df1.select_dtypes(include='object').columns.tolist()
ohe1 = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_ohe1 = ohe1.fit_transform(df1[categorical_cols1])

ohe1_cols = ohe1.get_feature_names_out(categorical_cols1)

df1_ohe = pd.DataFrame(X_ohe1, columns=ohe1_cols, index=df1.index)

df1_final = pd.concat([df1.drop(columns=categorical_cols1), df1_ohe], axis=1)
```

```python
df_final.to_csv('머신러닝_2013~2016FA.csv', index = False)
df1_final.to_csv('머신러닝_2017~2024FA.csv', index = False)
```
