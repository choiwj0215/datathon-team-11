```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("model/regression_prediction_4th_knn_tuning.csv")
```

```python
df1 = pd.read_csv("model/regression_prediction_4th_hardvoting_tuning.csv")
```

```python
df2 = pd.read_csv("model/regression_prediction_4th_xgb_tuning.csv")
```

```python
a1 = pd.read_csv('머신러닝_4차_전체_선수명포함.csv')
```

```python
a1 = a1[a1['FA년도'].isin([2023, 2024])]
a1 = a1.reset_index()
a1
```

```python
a1 = a1[['선수명', '1년당 계약 총액']]
a1
```

```python
df = pd.concat([df,a1], axis = 1)
```

```python
df1 = pd.concat([df1,a1], axis = 1)
```

```python
df2 = pd.concat([df2,a1], axis = 1)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
# print(df1['1년당 계약 총액'])
```

```python
# KNN_Tuning
player_fa_avg = df.groupby('선수명')['1년당 계약 총액'].mean().reset_index()

player_fa_avg
```

```python
# hardvoting_tuning
player_fa_avg = df1.groupby('선수명')['1년당 계약 총액'].mean().reset_index()

player_fa_avg
```

```python
# xgb_tuning
player_fa_avg = df2.groupby('선수명')['1년당 계약 총액'].mean().reset_index()

player_fa_avg
```
