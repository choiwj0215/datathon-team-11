```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("model/regression_prediction_choice_2016_gb_basic.csv")
```

```python
df1 = pd.read_csv("model/regression_prediction_choice_2024_gb_tuning.csv")
```

```python
df3 = pd.read_csv("model/regression_prediction_choice_2016_hardvoting_tuning.csv")
```

```python
df4 = pd.read_csv("model/regression_prediction_choice_2024_knn_tuning.csv")
```

```python
a1 = pd.read_csv('2026_FA예정_타자.csv')
```

```python
a1 = a1['선수명']
```

```python
df = pd.concat([df,a1], axis = 1)
```

```python
df1 = pd.concat([df1,a1], axis = 1)
```

```python
df3 = pd.concat([df3,a1], axis = 1)
```

```python
df4 = pd.concat([df4,a1], axis = 1)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
# print(df1['1년당 계약 총액'])
```

```python
# ~2016년 단순평균  gb_basic
player_fa_avg = df.groupby('선수명')['1년당 계약 총액'].mean().reset_index()

player_fa_avg
```

```python
# ~2016 년도별 가중치 평균 gb_basic
latest_year = df.groupby('선수명')['플레이년도'].transform('max')

# 최신 연도 대비 몇 년 전인지 계산
df['연도차'] = latest_year - df['플레이년도'] + 1

# 가중치 딕셔너리
weights = {1: 0.4, 2: 0.4, 3: 0.3, 4: 0.2, 5:0.1}  # 최신 연도일수록 가중치 높음

df['가중치'] = df['연도차'].map(weights)

# 가중평균 FA 계약 총액 계산
player_fa_weighted_avg = (
    df.groupby('선수명').apply(
        lambda x: np.average(x['1년당 계약 총액'], weights=x['가중치'])
    ).reset_index(name='가중평균 1년당 계약 총액')
)

player_fa_weighted_avg
```
