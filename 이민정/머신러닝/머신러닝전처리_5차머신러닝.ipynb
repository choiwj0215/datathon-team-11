```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
df= pd.read_csv('2017~2024_FA_4년정보.csv')
```

```python
df.columns
```

```python
df = df[['FA년도','선수명_수정', '포지션', '타',  'FA등급', '플레이년도', '나이',
       '게임수', '타석수', '타수', '득점', '안타', '2루타','사구', '홈런', '타점','볼넷', '삼진', '병살타', '공격WAR', '연봉',
       'wRC+', '장타율', '출루율', 'OPS', '구단지역', '국대출전횟수', '원클럽맨여부', '타격 WPA', 'WPA+',
       'WPA-', 'WPA', '홈런기여점수', '메인 타순', '1년당 계약 총액']]
```

```python
df['메인 타순'] = df['메인 타순'].str.extract('(\d+)').astype(float)
df['선수명'] = df['선수명_수정']
df = df.drop('선수명_수정', axis = 1)
```

```python
df = df.drop('선수명', axis=1)
```

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_ohe = ohe.fit_transform(df[categorical_cols])

ohe_cols = ohe.get_feature_names_out(categorical_cols)

df_ohe = pd.DataFrame(X_ohe, columns=ohe_cols, index=df.index)

df_final = pd.concat([df.drop(columns=categorical_cols), df_ohe], axis=1)
```

```python
df_final
```

```python
a1 = df_final[df_final['FA년도'] < 2024]
a2 = df_final[df_final['FA년도'] >= 2024]
```

```python
a1.reset_index()
```

```python
a2.reset_index()
```

```python
a1.to_csv('머신러닝_5차_2017~2023.csv', index = False)
a2.to_csv('머신러닝_5차_2024.csv', index = False)
```
