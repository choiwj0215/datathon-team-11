```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('2026_FA예정_타자.csv')
df
```

```python
df.columns
```

```python
df = df.drop(['구단명', '선수명','잔류 여부'], axis = 1)
```

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
ohe = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')

X_ohe = ohe.fit_transform(df[categorical_cols])

ohe_cols = ohe.get_feature_names_out(categorical_cols)

df_ohe = pd.DataFrame(X_ohe, columns=ohe_cols, index=df.index)

df_final = pd.concat([df.drop(columns=categorical_cols), df_ohe], axis=1)
```

```python
pd.set_option('display.max_columns', None)
df_final
```

```python
df_final.to_csv('머신러닝_2026_FA예정_타자.csv', index = False)
```
