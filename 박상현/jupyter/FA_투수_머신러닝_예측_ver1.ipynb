```python
# 1. 불러오기
test = pd.read_csv(f'{root_path}/2026_FA예정_투수.csv', encoding='utf-8')
train = pd.read_csv(f'{root_path}/train_선수요약_AAV.csv', encoding='utf-8')
```

```python
# 2. train 결측치 제거
X = train.drop(columns=['선수명', 'AAV']).dropna()
y = train.loc[X.index, 'AAV']
```

```python
# 3. 모델 학습
model = RandomForestRegressor(random_state=42)
model.fit(X, y)
```

```python
# 5. test 필요없는 컬럼 제거
drop_cols = [
    'FA년도', '구단명', 'FA등급', '비고1', '구분',
    '잔류 여부', '구종1', '구종2', '구종1 구종가치',
    'FA 계약 총액', 'FA 계약 연수'
]
test = test.drop(columns=drop_cols, errors='ignore')
```

```python
# 6. 선수별 평균 요약
test_grouped = test.groupby("선수명").mean(numeric_only=True).reset_index()
```

```python
# 7. 범주형 인코딩 ('투', '세부 포지션')
cat_cols = ['투', '세부 포지션']
for col in cat_cols:
    if col in test.columns:
        dummies = pd.get_dummies(test[[col, '선수명']], columns=[col])
        dummies = dummies.groupby("선수명").max().reset_index()
        test_grouped = pd.merge(test_grouped, dummies, on="선수명", how="left")
```

```python
# 8. train 기준으로 피처 맞추기
X_train_cols = X.columns
for col in X_train_cols:
    if col not in test_grouped.columns:
        test_grouped[col] = 0
```

```python
# 컬럼 순서 정렬
X_test = test_grouped[X_train_cols]
```

```python
# 9. 예측
test_preds = model.predict(X_test)
```

```python
# 10. 결과 정리
result = test_grouped[['선수명']].copy()
result['예측_AAV'] = test_preds.round().astype(int)
```

```python
# 결과 확인
print(result)
```
